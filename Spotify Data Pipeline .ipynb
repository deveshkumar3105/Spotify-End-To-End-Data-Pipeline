```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
```

```python
client_credentials_manager = SpotifyClientCredentials(client_id="",client_secret="")
```

```python
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

```python
playlist_link = "https://open.spotify.com/playlist/37i9dQZEVXbLZ52XmnySJg"
```

```python
playlist_link.split("/")
```

```python
playlist_URI = playlist_link.split("/")[-1]
```

```python
playlist_URI
```

```python
sp.playlist_tracks(playlist_URI)
```

```python
data = sp.playlist_tracks(playlist_URI)
```

```python
data['items'][0]['track']['album']['id']
data['items'][0]['track']['album']['name']
data['items'][0]['track']['album']['release_date']
data['items'][0]['track']['album']['total_tracks']
data['items'][0]['track']['album']['external_urls']['spotify']
```

```python
album_list = []
for row in data['items']:
    album_id = row['track']['album']['id']
    album_name = row['track']['album']['name']
    album_release_date = row['track']['album']['release_date']
    album_total_tracks = row['track']['album']['total_tracks']
    album_url = row['track']['album']['external_urls']['spotify']
    
    album_items ={'album_id':album_id,'album_name':album_name,'album_release_date':album_release_date,'album_total_tracks':album_total_tracks,'album_url':album_url}
    
    album_list.append(album_items)
```

```python
artist_list = []
for row in data['items']:
    for key,value in row.items():
        if key == 'track':
            for artist in value['artists']:
                artist_dict = {'artist_id':artist['id'],'artist_name':artist['name'],'external_urls':artist['href']}
                artist_list.append(artist_dict)
```

```python
song_list = []
for row in data['items']:
    song_id = row['track']['id']
    song_name = row['track']['name']
    song_duration = row ['track']['duration_ms']
    song_url = row['track']['external_urls']['spotify']
    song_popularity = row['track']['popularity']
    song_added = row['added_at']
    album_id = row['track']['album']['id']
    artist_id = row['track']['album']['artists'][0]['id']
    
    
    song_dict = {'song_id':song_id,'song_name':song_name,'song_duration':song_duration,'song_url':song_url,'song_popularity':song_popularity,'song_added':song_added,'album_id':album_id,'artist_id':artist_id}
    song_list.append(song_dict)
```

```python
# Album DataFrame
album_df = pd.DataFrame.from_dict(album_list)
```

```python
album_df.head()
```

```python
album_df = album_df.drop_duplicates(subset=['album_id'])
```

```python
# Artist DataFrame
artist_df = pd.DataFrame.from_dict(artist_list)
```

```python
artist_df.head()
```

```python
artist_df = artist_df.drop_duplicates(subset=['artist_id'])
```

```python
# Song DataFrame
song_df = pd.DataFrame.from_dict(song_list)
```

```python
song_df.head()
```

```python
album_df['album_release_date'] = pd.to_datetime(album_df['album_release_date'])
```

```python
song_df['song_added'] = pd.to_datetime(song_df['song_added'])
```
